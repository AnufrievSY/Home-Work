##Задание 4

```python
def equations(p):
  x, y = p
  return (1-(np.exp(x) + x*(1-y)), y - ((x**2)-1))
  
x_list = []
y_list = []

for i in range(-100, 100):
  x1, y1 = fsolve(equations, (i, i))
  x_list.append(round(x1, 2))
  y_list.append(round(y1, 2))
result_x = list(set(x_list))
result_y = list(set(y_list))

result_list = [[result_x[i], result_y[i]] for i in range(len(result_x))]

for i in result_list:
  print(i)

```

```python
def eq(p):
  x, y = p
  return (y - ((x**2)-1), (np.exp(x) + x*(1-y))-1)
x_list = []
y_list = []
for i in range(0, 100):
  x1, y1 = fsolve(eq, (i, i))
  if x1 > 0 and y1 > 0:
    x_list.append(round(x1, 1))
    y_list.append(round(y1, 1))
result_x = list(set(x_list))
result_y = list(set(y_list))
result_list = [[result_x[i], result_y[i]] for i in range(len(result_x))]
for i in result_list:
  print(i)
```
