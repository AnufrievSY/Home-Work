```python
b = randint(1,10)  # длинна иглы
alfa = randint(0, 360)  # угол под которым лежит игла
x0 = 10  # координаты нижней точки иглы
y0 = 10 # координаты нижней точки иглы
x_ = np.linspace(-20, 20, 41)
a = randint(1, 5)  # расстояние между линиями
x_list = []
y_list = []

"""Отрисовка иглы"""
x1 = np.sin(alfa)*b
y1 = np.cos(alfa)*b

x = np.linspace(x0, x1, 20)
y = np.linspace(y0, y1, 20)
plt.plot(x, y, color = 'black')

"""Отрисовка линий"""
for i in range(round(21/a)):
  y_ = 0*x_+i*a
  plt.plot(x_, y_, color='cyan')
  x_list.append(set(list(x_)))
  y_list.append(set(list(y_)))
  
"""Поиск точек которые пересекает игла"""
list_ = []
for i in y_list:
  for i_ in i:
    list_.append(int(i_))

intersection_list = []
for i in y:
  try:
    list_.index(int(i))
    intersection_list.append(int(i))
  except ValueError:
    None
print(f'Иглай пересекает Y: {set(intersection_list)}')
```
