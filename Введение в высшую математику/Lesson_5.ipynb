#Задание 3.1

```python
# Допустим значение успешного испытания:
x_value = randint(0, 10)

# Проводятся испытания, в котором получается опредленное значение:
def y_value():
  return randint(0, 10)

# Затем уточняется успешно ли проведенное испытание
def result():
  if x_value == y_value():
    return 1
  else:
    return 0

# Испытания проводятся 'k' колличество раз
n = 10

# Составляется список из успешных и провальных испытаний:
result_list = [result() for _ in range(n)]
print('Последовательность независимых испытаний, где:\n',
      '0 - провальное испытание\n',
      '1 - успешное испытание\n',
      result_list, '\n')

# Таким образом, испытания были успешны 'n' раз 
k = result_list.count(1)

# Тогда биномиальный коэффициент будет равен:
C = math.factorial(n)/(math.factorial(n-k)*math.factorial(k))

# Шанс успеха каждого из испытаний равен:
p = 1/n

# Вероятной противоположного события равна:
q = 1-p

# Формула Бернули:
P = C*(p**k)*(q**(n-k))
print(f'Вероятность такого результата или вероятность того что в {n} независимых\n'
      f'испытаниях, в каждом из которых вероятность появления события равна {p},\n'
      f'событие наступит ровно {k} раз:\n',
      P)
```

#Задание 3.2

```python
for i in range(1, 6):
  n = randint(0,100)  #колличество испытаний
  result_list = [randint(0, 1) for i in range(n)]
  k = result_list.count(1)  #колличество успешных испытаний
  С = math.factorial(n)/(math.factorial(n-k)*math.factorial(k))  #биномиальный коэффициент
  p = 50/100 # шанс успеха каждого из испытаний
  q = 1-p  # вероятной противоположного события
  P = C*(p**k)*(q**(n-k)) # формула Бернули

  print(f'Вероятность того что в {n} независимых испытаниях, в каждом из которых\n'
        f'вероятность появления события равна {p*100}%, событие наступит ровно {k} раз:\n',
        P,'\n','--------------------------------------------------------------------')
```
