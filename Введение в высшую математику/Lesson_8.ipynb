```python
import numpy as np
from scipy.optimize import fsolve
```

#Задание 3

```python
def equations(p):
    x, y = p
    f_1 = (x**2) - (y**2) + (3*x*(y**3)) - (2*(x**2)*(y**2)) +(2*x) - 5
    f_2 = (3*(y**3)) - (2*(x**2)) + (2*(x**3)*y) - (5*(x**2)*(y**2)) + 5
    return (f_1, f_2)

result = []
for x0 in np.arange(-5, 5, 0.0001):
  (x, y), info, ier, mesg =  fsolve(equations, (x0, x0), full_output=True)
  if ier == 1:
    x = round(float(x), 6)
    y = round(float(y), 6)
    try:
      result.index([x, y])
    except ValueError:
      result.append([x, y])

for i in result:
  print(i)
```
