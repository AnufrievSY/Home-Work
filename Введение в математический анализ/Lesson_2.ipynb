```python
def check(x_list):
  i = 1
  x_1 = 1
  x_2 = 0
  while True:
    x_1 = ''.join(x_list[0:i])
    x_2= ''.join(x_list[0+i:i+i])
    i += 1
    if i == 100:
      return f"0.{''.join(x_list)}"
      break
    elif x_1 == x_2:
      return f'0.({x_1})'
      break

def answer(x, number_system):
  x_list = []
  i = 0
  while True:
    x = x*number_system
    x_list.append(str(round(x//1)))
    x = x%1
    i += 1
    if x == 0:
      return check(x_list)
      break
    elif i == 100:
      return check(x_list)
      break

number_system = 0
while True:
  number = 2/12 #изначальное число
  number_system += 1 #система счисления
  find = '0.(13)' #число которое мы ищем
  if answer(number, number_system) == find:
    print(f'Чтобы из 2/12 получилось число {find} необходимо перевести его в {number_system}-ичную систему счиления.')
    break
  elif number_system == 1000:
    print('not answer')
    break
  
  
```
