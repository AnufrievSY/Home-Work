##Задание 2

```python
from random import randint

while True:
  y1 , y2, y3 = [randint(1, 100) for i in range(3)]
  result = (1/y1)+(1/y2)+(1/y3)
  if  result == 1 and y1 != y2 and y1 != y3 and y2 != y3:
    break
print(f'1/{y1} + 1/{y2} + 1/{y3} = {result}')
```

##Задание 3

```python
while True:
  y = list(set([randint(1, 100) for i in range(6)]))
  if len(y) == 6:
    result = sum([1/i for i in y])
    if result == 1:
      print(f'1/{y[0]} + 1/{y[1]} + 1/{y[2]} + 1/{y[3]} + 1/{y[4]} + 1/{y[5]} = {result}')
      break
```
